```python
import numpy as np
```

```python
def w(x, list_x):
    wx = 1
    for xi in list_x:
        wx = wx * (x - xi)
    return wx


def dw(k, list_x):
    xk = list_x[k]
    new_list = list_x[:k] + list_x[k+1:]
    dwx = w(xk, new_list)
    return dwx


def lk(x, k, list_x):
    xk = list_x[k]
    wx = w(x, list_x)
    dwxk = dw(k, list_x)
    l = wx / (dwxk * (x - xk))
    print(f"k={k}, x={x:.6f}, wx={wx:.6f}, dwxk={dwxk:.6f}, xk={xk:.6f}, l={l:.6f}")
    return l


def lagrange_interpolation(list_x, list_y, n=1):
    # we assume list_x is sorted in increasing order
    assert n <= len(list_x) and len(list_x) == len(list_y)
    
    def L(x):
        start = -1
        for i, xi in enumerate(list_x):
            if xi > x:
                start = i - 1
                if start < 0:
                    start = 0
                break

        assert start+n <= len(list_x)
        
        target_list_x = list_x[start:start+n+1]
        target_list_y = list_y[start:start+n+1]
        print(f"start={start}")
        print(f"target_list_x={target_list_x}")
        print(f"target_list_y={target_list_y}")
        
        Lx = 0
        for k, yk in enumerate(target_list_y):
            l = lk(x, k, target_list_x)
            Lx = Lx + yk * l
            print(f"Lx={Lx:.6f}, l={l:.6f}, yk={yk:.6f}")
        return Lx

    return L
```
