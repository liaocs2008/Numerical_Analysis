### x^2 + x - 1 = 0

x = (sqrt(5)-1)/2

x(x+1) = 1

x = 1 / (x+1)

```python
def iterative_method(x0, n):
    for _ in range(n):
        x1 = 1 / (x0 + 1)

        # update
        x0 = x1
    return x1
```

```python
x_true = (np.sqrt(5)-1)/2
x_true
```

```python
x5 = iterative_method(1, 5)
x5
```

```python
x10 = iterative_method(1, 10)
x10
```

x1 = 1 / (x0 + 1)

new_x1 = x1 + w * (x1 - x0)

```python
def relaxed_iterative_method(x0, n, w):
    for _ in range(n):
        x1 = 1 / (x0 + 1)
        new_x1 = x1 + w * (x1 - x0)

        # update
        x0 = new_x1
    return x1
```

```python
x5 = relaxed_iterative_method(1, 5, -7/25)  # faster convergence
x5
```

```python
x10 = relaxed_iterative_method(1, 10, -7/25)
x10
```

### I(0,1) = sum [f(xi_1) + f(xi)] * h/2, f(x) = e^x

```python
def f(x):
    e = 2.71828
    fx = e**x
    return fx
```

```python
def integral(f, a, b, n=100):
    h = (b - a) / n
    int_f = 0
    xi_1 = a
    for i in range(1, n):
        xi = a + i*h
        int_f += (f(xi) + f(xi_1)) * h/2
        
        # update
        xi_1 = xi

    return int_f
```

```python
integral(f, 0, 1, 1000)   # e^1 - e^0 = 1.71828...
```

```python
integral(f, 0, 1, 100)
```

```python
integral(f, 0, 1, 10)
```
