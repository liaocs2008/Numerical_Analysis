```python
import numpy as np
```

```python
def f0_k(list_x, list_y):
    k = len(list_x)
    if k == 1:
        return list_y[0]
        
    xk = list_x[-1]
    x0 = list_x[0]
    f1 = f0_k(list_x[1:k], list_y[1:k])
    f2 = f0_k(list_x[0:k-1], list_y[0:k-1])
    coef = (f1 - f2) / (xk - x0)
    print(f"f1={f1:.6f}, f2={f2:.6f}, xk={xk:.6f}, x0={x0:.6f}, coef={coef:.6f}")
    return coef
```

```python
def newton_interpolation(list_x, list_y, topk=None):
    n = len(list_x)
    if topk is None:
        topk = n
    
    coef_list = []
    for k in range(2,topk+1):  # # k can be at most n
        coef = f0_k(list_x[:k], list_y[:k])
        print(f"k={k}, coef={coef:.6f}")
        coef_list.append(coef)

    def f(x):
        fx = list_y[0]
        prod = 1
        print(f"coef_list={coef_list}")
        for k, coef in enumerate(coef_list):
            print(f"k={k}, coef={coef:.6f}, fx={fx}")
            prod = prod * (x - list_x[k])
            fx = fx + coef * prod
        return fx

    return f
```

```python
list_x = [0.4, 0.55, 0.65, 0.8, 0.9, 1.05]
list_y = [0.41075, 0.57815, 0.69675, 0.88811, 1.02652, 1.25382]
```

```python
f = newton_interpolation(list_x, list_y)
```

```python
fx = f(0.596)
print(f"fx={fx:.6f}")
```

```python
f = newton_interpolation(list_x, list_y, 5)
```

```python
fx = f(0.596)
print(f"fx={fx:.6f}")
```

### equal step size

```python
def newton_interpolation_with_equal_step(x0, h, list_y, topk=None):
    n = len(list_y)

    if topk is None:
        topk = n

    coef_list = []
    new_list_y = list_y.copy()
    for k in range(topk-1):
        new_list_y = [new_list_y[i] - new_list_y[i-1] for i in range(1,len(new_list_y))]
        # update
        coef = new_list_y[0]
        coef_list.append(coef)
        print(f"k={k}, coef={coef:.6f}")
        

    def f(x):
        fx = list_y[0]
        t = (x - x0) / h
        factorial = 1
        prod = 1
        print(f"coef_list={coef_list}")
        for k, coef in enumerate(coef_list):
            print(f"k={k}, t={t:.6f}, coef={coef:.6f}, fx={fx:.6f}")
            prod = prod * t / factorial
            fx = fx + coef * prod
            # update
            t = t - 1
            factorial = factorial + 1
        return fx

    return f
```

```python
x0 = 0
h = 0.1
list_y = [1, 0.995, 0.98007, 0.95534, 0.92106, 0.87758]
```

```python
f4 = newton_interpolation_with_equal_step(x0, h, list_y, 4)
```
